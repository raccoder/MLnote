```python
import numpy as np
import matplotlib.pyplot as plt
class MultiLinearRegression:
    def __init__(self):
        self.w = None
#         インスタンスを初期化
        
    def fit(self,X,y):
        X = np.hstack([np.ones((len(X),1)),X])
        self.w = np.linalg.inv(X.T@X)@X.T@y
#         デザイン行列を用いて学習を行わせる
        
    def predict(self,X_test):
        X_test = np.hstack([np.ones((len(X_test),1)),X_test])
        return X_test @ self.w
```

```python
# 合成データセットの生成
from sklearn.metrics import mean_squared_error

np.random.seed(0)
X = np.random.rand(100, 2)
true_coefficients = np.array([4, 3, 2])  # 実際の係数（切片含む）
y = np.dot(np.hstack([np.ones((len(X), 1)), X]), true_coefficients) + np.random.randn(100) * 0.5

# モデルの訓練
model = MultiLinearRegression()
model.fit(X, y)

# 予測性能の評価
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)

# 学習データと予測結果のプロット
# 注意: 2つの独立変数があるため、3Dプロットを使用
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 学習データのプロット
ax.scatter(X[:, 0], X[:, 1], y, color='blue', label='Training data')

# 予測結果のプロット
# メッシュグリッド上で予測を行う
x0_range = np.linspace(X[:, 0].min(), X[:, 0].max(), 20)
x1_range = np.linspace(X[:, 1].min(), X[:, 1].max(), 20)
x0, x1 = np.meshgrid(x0_range, x1_range)
X_grid = np.column_stack((x0.ravel(), x1.ravel()))
y_grid_pred = model.predict(X_grid).reshape(x0.shape)

# 予測面のプロット
ax.plot_surface(x0, x1, y_grid_pred, color='red', alpha=0.3, label='Prediction surface')

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')
plt.title('Training Data and Prediction Surface')
plt.show()

```
